```python
import h5py as hdf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import glob 
import seaborn as sns
#import netCDF as cdf 
%pylab inline
```

```python
files = ['Residuals/res400_erik.npy','Residuals/res600_erik.npy','Residuals/res700_erik.npy',
         'Residuals/res800_erik.npy','Residuals/resAll_erik.npy', # with erik's ref
         
         'Residuals/res400_400refset.npy','Residuals/res600_400refset.npy','Residuals/res700_400refset.npy',
         'Residuals/res800_400refset.npy','Residuals/resAllhpa_400refset.npy', # 400 refset
         
         'Residuals/res400_600refset.npy','Residuals/res600_600refset.npy','Residuals/res700_600refset.npy',
         'Residuals/res800_600refset.npy','Residuals/resAllhpa_600refset.npy', # 600 refset
         
         'Residuals/res400_700refset.npy','Residuals/res600_700refset.npy','Residuals/res700_700refset.npy',
         'Residuals/res800_700refset.npy','Residuals/resAllhpa_700refset.npy',  # 700 refset
         
         'Residuals/res400_800refset.npy','Residuals/res600_800refset.npy','Residuals/res700_800refset.npy',
         'Residuals/res800_800refset.npy','Residuals/resAllhpa_800refset.npy', # 800 refset
         
         'Residuals/res400_allrefset.npy','Residuals/res600_allrefset.npy','Residuals/res700_allrefset.npy',
         'Residuals/res800_allrefset.npy','Residuals/resAllhpa_allrefset.npy'  # All refset
        ]

residual_array = []

for i in range(len(files)):
    residual_array.append(np.load(files[i]))

#print len(residual_array),residual_array
```

```python
data    = hdf.File('SIFTER2/PCs_noise/1000hpa_SIFTER2_M02_EOF_2009_01','r') 
wav     = np.array(data['wavelength'])
wav.shape
```

```python
f        = plt.figure(1,figsize = (15,12))

rand_ret = randint(0,1000)
labels   = ['400hpa','600hpa','700hpa','800hpa','AllHpa']

for k,i in enumerate(residual_array):
    #print k,i
    
    if 0<=k<=4:
        plt.subplot(321)
        plt.plot(wav,residual_array[k][rand_ret],label=labels[k] + ' Retrieval # %i'%rand_ret)
        plt.axhline(0,linestyle='--',color='k')
        plt.ylabel('Residuals [-]')
        plt.title('Retrievals with Erik\'s Refset')
        
    if 5<=k<=9:
        plt.subplot(322)
        plt.plot(wav,residual_array[k][rand_ret])
        plt.axhline(0,linestyle='--',color='k')
        plt.ylabel('Residuals [-]')
        plt.title('Retrievals with 400 hpa Refset')
    
    if 10<=k<=14:
        plt.subplot(323)
        plt.plot(wav,residual_array[k][rand_ret])
        plt.axhline(0,linestyle='--',color='k')
        plt.ylabel('Residuals [-]')
        plt.title('Retrievals with 600 hpa Refset')
    
    if 15<=k<=19:
        plt.subplot(324)
        plt.plot(wav,residual_array[k][rand_ret])
        plt.axhline(0,linestyle='--',color='k')
        plt.ylabel('Residuals [-]')
        plt.title('Retrievals with 700 hpa Refset')
    
    if 20<=k<=24:
        plt.subplot(325)
        plt.plot(wav,residual_array[k][rand_ret])
        plt.axhline(0,linestyle='--',color='k')
        plt.ylabel('Residuals [-]')
        plt.title('Retrievals with 800 hpa Refset')

    if 25<=k<=29:
        plt.subplot(326)
        plt.plot(wav,residual_array[k][rand_ret])
        plt.axhline(0,linestyle='--',color='k')
        plt.ylabel('Residuals [-]')
        plt.title('Retrievals with All hpa Refset')

plt.subplot(321)
#plt.legend(bbox_to_anchor=(1.8,-1.5),fontsize=15)
plt.legend()
plt.show()
f.savefig('/usr/people/botiabo/Notebooks_Scripts/Report/ResidualsDiffRefsets.png',bbox_inches='tight',DPI=1000)
```
